```python
f=0                  # f 변수 초기화
try:                 # try문 시작
    f=open("LICENSE.txt","rt") # "LICENSE.txt" 파일을 텍스트 모드로 열기
    while True: # 무한 루프
        r=f.readline() # 파일에서 한 줄씩 읽기
        if not r: # 더 이상 읽어올 줄이 없으면
            break # 무한 루프 탈출
        print(r) # 한 줄씩 출력
except FileNotFoundError: # 파일이 없을 경우
    print("파일없음") # "파일없음" 출력
finally: # try 문이 끝나면 무조건 실행되는 부분
    if(f): f.close() # 파일 닫기
    print("프로그램 끝") # "프로그램 끝" 출력
```

```python
f=0                          # 파일 핸들링을 위한 변수 f 초기화
f=open("score.txt","rt")     # score.txt 파일을 텍스트 읽기 모드로 열기
while True:                  # 무한 루프 시작
    one_person=f.readline() # 파일에서 한 줄씩 읽기
    if not one_person:       # 읽을 내용이 없으면 루프 탈출
        break
    one_person=one_person.replace("\n","")  # 읽어들인 문자열에서 줄바꿈 문자를 없앰
    item=one_person.split(",")  # 읽어들인 문자열을 콤마를 기준으로 나눠서 item 리스트에 저장
    print("="*10)             # 구분을 위한 구분선 출력
    print("이름: ", item[0])   # item 리스트의 첫번째 항목은 이름
    print("국어: ", item[1])   # item 리스트의 두번째 항목은 국어 점수
    print("영어: ", item[2])   # item 리스트의 세번째 항목은 영어 점수
    print("수학: ", item[3])   # item 리스트의 네번째 항목은 수학 점수
    print("="*10)             # 구분을 위한 구분선 출력
f.close()                     # 파일 핸들링 종료
```

```python
import sqlite3 as s # sqlite3 라이브러리를 s라는 이름으로 import합니다.
con=s.connect("mydata.db") # mydata.db라는 데이터베이스 파일과 연결
cursor = con.cursor() # mydata.db라는 데이터베이스 파일과 연결

cursor.execute("drop table examtable;") #table 삭제
con.commit() # mydata.db라는 데이터베이스 파일과 연결

# examtable이라는 테이블을 생성
cursor.execute(" create table examtable( name varchar(20),studentid int not null primary key,kor int,eng int,mat int) ")
               
con.commit() #버퍼에 처리된 내용을 기록 commit
cursor.close() # 커서 객체를 닫는다
con.close() # 데이터베이스 연결을 닫는다
```

```python
import sqlite3 as s

con = s.connect("mydata.db") #데이터베이스 연결
cursor = con.cursor()

cursor.execute("select * from examtable") #examtable에서 모든 레코드를 가져옴

one_rec = cursor.fetchall() #가져온 레코드를 모두 변수 one_rec에 저장

for field in one_rec: #반복문을 통해 one_rec에 저장된 각 레코드를 순회
    kor = field[2] #각 레코드의 국어 성적
    eng = field[3] #각 레코드의 영어 성적
    mat = field[4] #각 레코드의 수학 성적
    total = kor + eng + mat #총점 계산
    avg = int(total / 3) #평균 계산
    print("이름:%s 학번:%s 국어:%3d 영어:%3d 수학:%3d 총점:%3d 평균:%3d" % (field[0], field[1], kor, eng, mat, total, avg))

cursor.close() #커서 닫기
con.close() #데이터베이스 연결 닫기
```
